```python
import pandas as pd
import altair as alt
import matplotlib.pyplot as plt
import seaborn as sns
```

## Dataset
I obtained the dataset from Kaggle. https://www.kaggle.com/datasets/nelgiriyewithana/countries-of-the-world-2023

### Global Country Information Dataset 2023
A Comprehensive Dataset Empowering In-Depth Analysis and Cross-Country Insights
### Description:
This comprehensive dataset provides a wealth of information about all countries worldwide, covering a wide range of indicators and attributes. It encompasses demographic statistics, economic indicators, environmental factors, healthcare metrics, education statistics, and much more. With every country represented, this dataset offers a complete global perspective on various aspects of nations, enabling in-depth analyses and cross-country comparisons.

The dataset has 35 features of 195 countries' information.
- Country: Name of the country.
- Density (P/Km2): Population density measured in persons per square kilometer.
- Abbreviation: Abbreviation or code representing the country.
- Agricultural Land (%): Percentage of land area used for agricultural purposes.
- Land Area (Km2): Total land area of the country in square kilometers.
- Armed Forces Size: Size of the armed forces in the country.
- Birth Rate: Number of births per 1,000 population per year.
- Calling Code: International calling code for the country.
- Capital/Major City: Name of the capital or major city.
- CO2 Emissions: Carbon dioxide emissions in tons.
- CPI: Consumer Price Index, a measure of inflation and purchasing power.
- CPI Change (%): Percentage change in the Consumer Price Index compared to the previous year.
- Currency_Code: Currency code used in the country.
- Fertility Rate: Average number of children born to a woman during her lifetime.
- Forested Area (%): Percentage of land area covered by forests.
- Gasoline_Price: Price of gasoline per liter in local currency.
- GDP: Gross Domestic Product, the total value of goods and services produced in the country.
- Gross Primary Education Enrollment (%): Gross enrollment ratio for primary education.
- Gross Tertiary Education Enrollment (%): Gross enrollment ratio for tertiary education.
- Infant Mortality: Number of deaths per 1,000 live births before reaching one year of age.
- Largest City: Name of the country's largest city.
- Life Expectancy: Average number of years a newborn is expected to live.
- Maternal Mortality Ratio: Number of maternal deaths per 100,000 live births.
- Minimum Wage: Minimum wage level in local currency.
- Official Language: Official language(s) spoken in the country.
- Out of Pocket Health Expenditure (%): Percentage of total health expenditure paid out-of-pocket by individuals.
- Physicians per Thousand: Number of physicians per thousand people.
- Population: Total population of the country.
- Population: Labor Force Participation (%): Percentage of the population that is part of the labor force.
- Tax Revenue (%): Tax revenue as a percentage of GDP.
- Total Tax Rate: Overall tax burden as a percentage of commercial profits.
- Unemployment Rate: Percentage of the labor force that is unemployed.
- Urban Population: Percentage of the population living in urban areas.
- Latitude: Latitude coordinate of the country's location.
- Longitude: Longitude coordinate of the country's location.

```python
df = pd.read_csv('data/world-data-2023.csv')
world_gdp = pd.read_csv('data/World GDP Dataset.csv')
df.head()
```

## Data Cleaning
I did the basic data cleaning like removing the missing and incomplete data. I needed to preprocess the columns to better compatible format.

```python
df.isnull().sum()
```

```python
world_gdp = world_gdp.dropna()
df = df[(df.Latitude.notnull()) & (df.Population.notnull()) & (df.GDP.notnull())]
df = df[df.Country != 'Eswatini']
```

```python
columns_to_convert = ['Density\n(P/Km2)', 'Agricultural Land( %)',
        'Land Area(Km2)', 'Armed Forces size', 'Birth Rate',
        'Co2-Emissions', 'CPI', 'CPI Change (%)',
        'Fertility Rate', 'Forested Area (%)',
        'Gasoline Price', 'GDP', 'Gross primary education enrollment (%)',
        'Gross tertiary education enrollment (%)', 'Infant mortality',
        'Life expectancy', 'Maternal mortality ratio',
        'Minimum wage', 'Out of pocket health expenditure',
        'Physicians per thousand', 'Population',
        'Population: Labor force participation (%)', 'Tax revenue (%)',
        'Total tax rate', 'Unemployment rate', 'Urban_population']

convert_to_float = lambda x: float(str(x).replace(",", "").replace("%", "").replace("$",""))

for column in columns_to_convert:
    df[column] = df[column].apply(convert_to_float)
```

```python
import json

continents = f = open('data/continents.json', "r")
continents = json.loads(f.read())
continents_info = pd.DataFrame(continents)

df = df.merge(continents_info, how = 'left', left_on = 'Country', right_on = 'country')
```

```python
df_chose = df.loc[:, ['Density\n(P/Km2)','Agricultural Land( %)',
       'Land Area(Km2)', 'Armed Forces size', 'Birth Rate',
       'Co2-Emissions', 'CPI', 'CPI Change (%)',
       'Fertility Rate', 'Forested Area (%)',
       'Gasoline Price', 'GDP', 'Gross primary education enrollment (%)',
       'Gross tertiary education enrollment (%)', 'Infant mortality',
       'Life expectancy', 'Maternal mortality ratio',
       'Minimum wage', 'Out of pocket health expenditure',
       'Physicians per thousand', 'Population',
       'Population: Labor force participation (%)', 'Tax revenue (%)',
       'Total tax rate', 'Unemployment rate', 'Urban_population']]
                      
correlation_matrix = df_chose.corr()
fig, ax = plt.subplots(figsize=(20,10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', ax=ax).set(title="Correlation between the features")
```

## World Population vs Life Expectancy

```python
import altair as alt
from vega_datasets import data

min_value = df['Population'].min()
max_value = df['Population'].max()
min_size = 10
max_size = 1000

countries = alt.topo_feature(data.world_110m.url, 'countries')

background = alt.Chart(countries).mark_geoshape(fill='lightgray', stroke = 'white').project('equirectangular').properties(
    width=1000,
    height=500,
    title='Population Map 2023'
)

points = alt.Chart(df).mark_circle().encode(
    longitude='Longitude',
    latitude='Latitude',
    size=alt.Size('Population', scale=alt.Scale(domain=[min_value, max_value], range=[min_size, max_size])),
    tooltip=['Country'],
    color='Country:N'
)

background + points
```

```python
alt.Chart(df).mark_circle(size=60).encode(
    x='Life expectancy',
    y='Population',
    color='Country',
    tooltip=['Country', 'Population', 'Life expectancy']
).properties(
    width=800,
    height=400
).interactive()
```

```python

source = df[(df['Country'] != 'India') & (df['Country'] != 'China')]

brush = alt.selection_interval()

points = alt.Chart(source).mark_point().encode(
    x='Population',
    y='Life expectancy',
    color=alt.condition(brush, 'Country', alt.value('lightgray'))
).add_params(
    brush
).properties(
    width=1000,
    height=400
)

bars = alt.Chart(source).mark_bar().encode(
    y='Population',
    color='Country',
    x='Life expectancy'
).transform_filter(
    brush
)

points & bars
```

## Economic Development

I want to explore how much correlation between GDP and population. Normally, the greater population should mean greater demand in goods and services so, we should generally see a positive correlation with GDP.

```python
feature_cols = ['Country', 'Density\n(P/Km2)', 'Population', 'GDP', 'Gross primary education enrollment (%)', 'Gross tertiary education enrollment (%)', 'Population: Labor force participation (%)', 'Unemployment rate', 'continent']
df_explore_1 = df.loc[:, feature_cols]
df_explore_1.head()
```

```python
sns.pairplot(df_explore_1, height=3)
```

```python
correlation_matrix = df_explore_1.loc[:, feature_cols[1:-1]].corr()
fig, ax = plt.subplots(figsize=(10,5))         # Sample figsize in inches

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', ax=ax)
```

As I expected, the Population and GDP has a correlation value of 0.63, which indicates a positive relationship between the two variables. This means that, in general, as one variable increases, the other tends to increase as well. 


In Education, the dataset show no significant correlation between primary education and GDP while tertiary education show some positive relationship.

Primary Education Enrollment & Labor Force Participation
A correlation of 0.22 indicates a weak positive relationship between primary education enrollment and labor force participation. This may be due to educated workforce leads to more participation.


```python
alt.Chart(df_explore_1).mark_circle(size=60).encode(
    x='Population',
    y='GDP',
    color='Country',
    tooltip=['Country', 'Population', 'GDP']
).properties(
    width=1000,
    height=500
).interactive()
```

```python
alt.Chart(df_explore_1).mark_circle(size=60).encode(
    x='GDP',
    y='Gross tertiary education enrollment (%)',
    color='Country',
    tooltip=['Country', 'GDP', 'Gross tertiary education enrollment (%)']
).properties(
    width=1000,
    height=500
).interactive()
```

```python
top_10_gdp_countries
```

Those 10 countries are the highest GDP earned in 2023. So, let's take a look at back to the history. I chose to explore another world GDP dataset and plot the 20th century record.

```python
world_gdp.rename(columns = {
    'GDP, current prices (Billions of U.S. dollars)': 'Country'
}, inplace = True)
```

```python
select_cols = ['Country', '2000', '2001', '2002', '2003', '2004',
       '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
       '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022',
       '2023']
```

```python
world_gdp = world_gdp.loc[:, select_cols]
```

```python
world_gdp
```

```python
def change_name(df):
    if df.Country == "China, People's Republic of":
        return "China"
    elif  df.Country == "Korea, Republic of":
        return "South Korea"
    else:
        return df.Country
world_gdp['Country'] = world_gdp.apply(change_name, axis = 1)
```

```python
top_10_countries_gdp = world_gdp.loc[world_gdp.Country.isin(top_10_gdp_countries.Country.values)]
```

```python
df_melted = pd.melt(top_10_countries_gdp, id_vars=['Country'], var_name='Year', value_name='Value')
df_melted['Year'] = df_melted['Year'].astype(int)
```

```python
alt.Chart(top_10_gdp_countries).mark_bar().encode(
    y='Country',
    x='GDP'
).properties(
    width=1000, height = 300,
    title="Top 10 Highest GDP Countries in 2023"
)
```

```python
alt.Chart(df_melted).mark_bar().encode(
    x='Year',
    y='Value',
    color='Country',
    column='Country'
).properties(
    title="Top 10 Highest GDP Countries [2020-2023]"
)
```

```python
alt.Chart(df_melted).mark_line().encode(
    x='Year',
    y='Value',
    color='Country'
).properties(
    width=1000,
    height=500
).interactive()
```

US has always higher GDP in 20th century. China GDP started growing in 2009 and exceeded Japan. 

```python
def unemployment_cat(df):
    if df['Unemployment rate'] < 5:
        return '< 5%'
    elif 5 < df['Unemployment rate'] < 10:
        return '< 10%'
    elif 10 < df['Unemployment rate'] < 15:
        return '< 15% '
    elif 15 < df['Unemployment rate'] < 20:
        return '< 20% '
    elif df['Unemployment rate'] > 20:
        return '> 20% '
    
```

```python
df['Unemployment Category'] = df.apply(unemployment_cat, axis = 1)
```

```python
df_pie = df[df['Unemployment Category'].notnull()]
```

```python
employment_summary = df.groupby('Unemployment Category').size().reset_index(name='Count')
```

```python
base = alt.Chart(employment_summary).encode(
    alt.Theta("Unemployment Category").stack(True),
    alt.Radius("Count").scale(type="sqrt", zero=True, rangeMin=25),
    color="Count",
).properties(
    width=1000,
    height=500
).properties(
    title="Pie Chart of Unemployment Rate"
)

c1 = base.mark_arc(innerRadius=20, stroke="#fff")

c2 = base.mark_text(radiusOffset=10).encode(text="Unemployment Category")

c1 + c2
```

## Environmental Sustainability Analysis

Environmental sustainability analysis is a critical area of study, especially in the context of climate change and natural resource management. I am going to find the correlation of feactures.

```python
df_env = df.loc[:, ['Density\n(P/Km2)','Agricultural Land( %)',
       'Land Area(Km2)', 'Armed Forces size',
       'Co2-Emissions', 'GDP',  'Forested Area (%)', 'Population', 'Urban_population']]

correlation_matrix = df_env.corr()
fig, ax = plt.subplots(figsize=(10,8))         # Sample figsize in inches

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', ax=ax)
```

CO2 Emissions & GDP (0.917): There's a very strong positive correlation between a country's CO2 emissions and its GDP. This makes sense because of the energy consumption tied to economic activity.

Population & CO2 Emissions (0.81): Urban population is also highly correlated with CO2 emissions, possibly due to higher energy consumption. More people needs more energy for household and public mobility, transportation can result in higher CO2 release.

## Healthcare Accessibility and Effectiveness

```python
df_health = df.loc[:, ['Birth Rate',
       'Fertility Rate',
       'Gross tertiary education enrollment (%)', 'Infant mortality',
       'Life expectancy', 'Maternal mortality ratio',
       'Minimum wage', 'Out of pocket health expenditure',
       'Physicians per thousand']]
                      
correlation_matrix = df_health.corr()
fig, ax = plt.subplots(figsize=(20,10))         # Sample figsize in inches

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', ax=ax)
```

```python
alt.Chart(df).mark_circle(size=60).encode(
    x='Physicians per thousand',
    y='Life expectancy',
    color='continent',
    tooltip=['continent', 'Physicians per thousand', 'Life expectancy']
).properties(
    width=1000,
    height=500
).interactive()
```

From this visualisation, we can see that the number of doctors are very important the longevity of the population. This visualisualisation use colors to tell the continents. So, we can clearly see that Africa needs more doctors and supports.

```python
alt.Chart(df).mark_circle(size=60).encode(
    x='Gross tertiary education enrollment (%)',
    y='Life expectancy',
    color='continent',
    tooltip=['continent', 'Gross tertiary education enrollment (%)', 'Life expectancy']
).properties(
    width=1000,
    height=500
).interactive()
```

Education is also important for longer life expectancy. BUT we should be mindful that correlation doesn't mean causation. So, it may not mean more educated people lives longer but it's interesting to see this correlation.

```python
alt.Chart(df).mark_circle(size=60).encode(
    x='Birth Rate',
    y='Life expectancy',
    tooltip=['Birth Rate', 'Life expectancy']
).properties(
    width=1000,
    height=500
).interactive()
```
